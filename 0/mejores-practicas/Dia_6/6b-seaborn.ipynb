## Introducción a Seaborn

Seaborn es una biblioteca de visualización de datos en Python basada en Matplotlib. Proporciona una interfaz de alto nivel para dibujar gráficos estadísticos atractivos e informativos. Seaborn te ayuda a explorar y entender tus datos. Funciona bien con las estructuras de datos de Pandas e integra de manera eficiente la biblioteca Matplotlib.

## Empezando con Seaborn

Primero, necesitamos importar nuestras bibliotecas. Usaremos Seaborn para la visualización, Matplotlib para personalizaciones adicionales, NumPy para operaciones numéricas y Pandas para la manipulación de datos.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Carguemos un conjunto de datos para nuestros ejemplos. Seaborn viene con algunos conjuntos de datos integrados, que son excelentes para aprender y practicar la visualización de datos.

```python
# Cargar el conjunto de datos 'tips' de Seaborn
tips = sns.load_dataset('tips')
```

## Gráficas Básicas en Seaborn

### Gráficas de Distribución

Las gráficas de distribución son útiles para examinar la distribución de una o más variables en un conjunto de datos. Comencemos con un simple histograma.

```python
sns.histplot(tips['total_bill'])
plt.title('Distribución de Cuentas Totales')
plt.show()
```

Ahora, visualicemos distribuciones bivariadas usando un gráfico de dispersión.

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Gráfico de Dispersión de Cuenta Total vs. Propina')
plt.show()
```

### Gráficas de Datos Categóricos

Los datos categóricos se pueden visualizar usando gráficos como los boxplots y violin plots, que muestran la distribución de una variable numérica para diferentes categorías en tus datos.

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Boxplot de Cuentas Totales por Día')
plt.show()
```

## Visualizaciones Avanzadas

### Mapas de Calor

Los mapas de calor son excelentes para visualizar matrices y pueden ser particularmente útiles para mostrar matrices de correlación.

```python
# cargar csv
housing_data = pd.read_csv('train.csv')
data_floats = housing_data.select_dtypes(include='float64')
data_floats.corr()

sns.heatmap(data_floats.corr());
plt.title('Matriz de Correlación del Conjunto de Datos Tips')
plt.show()
```

### Cuadrículas Facetadas

Las cuadrículas facetadas son útiles para crear una matriz de gráficos basada en una característica.

```python
# Cuadrícula Facetada de tiempo vs estado de fumador para la distribución de la cuenta total
g = sns.FacetGrid(tips, col="time", row="smoker")
g = g.map(plt.hist, "total_bill")
plt.show()
```

## Estilización y Personalización

Seaborn permite la estilización fácil de gráficos para hacerlos más visualmente atractivos.

```python
# Establecer el estilo estético de los gráficos
sns.set_style("whitegrid")

# Gráfico con el estilo actualizado
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Boxplot Estilizado de Cuentas Totales por Día')
plt.show()
```
