# Visualización de Datos con Matplotlib

Matplotlib es una poderosa biblioteca para crear visualizaciones estáticas, animadas e interactivas en Python. Proporciona una API orientada a objetos para incrustar gráficas en aplicaciones.

### Empezando con Matplotlib

Comencemos importando Matplotlib junto con NumPy, que usaremos para generar datos de muestra.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Conceptos Básicos de la Graficación

Graficar en Matplotlib es tan simple como llamar a la función `plot` con algunos datos, y luego llamar a `show()` para mostrar la figura.

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('Algunos números')
plt.show()
```

### Gráfica con Doble Eje Y

Útil para comparar dos conjuntos de datos con diferentes unidades o escalas.

```python
fig, ax1 = plt.subplots()

x = np.arange(0, 10, 0.5)
y1 = np.exp(x)
y2 = np.log(x + 1)

ax1.plot(x, y1, 'g-')
ax1.set_xlabel('Datos X')
ax1.set_ylabel('Exp', color='g')

ax2 = ax1.twinx()
ax2.plot(x, y2, 'b-')
ax2.set_ylabel('Log', color='b')

plt.title("Gráfica con Doble Eje Y")
plt.show()
```

### Gráfica de Dispersión con Línea de Tendencia

Ideal para mostrar la relación entre dos variables continuas.

```python
# Generando datos de muestra
x = np.random.random(100) * 10
y = 2.5 * x + np.random.normal(size=100)

# Creando gráfica de dispersión
plt.scatter(x, y)

# Calculando y graficando la línea de tendencia
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--")

plt.title("Gráfica de Dispersión con Línea de Tendencia")
plt.xlabel("Valor X")
plt.ylabel("Valor Y")
plt.show()
```

### Histograma con Curva de Distribución

Genial para visualizar la distribución de un conjunto de datos y entender sus propiedades estadísticas.

```python
# Generando datos de muestra
data = np.random.normal(loc=0, scale=1, size=1000)

# Creando histograma
n, bins, patches = plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Añadiendo una curva de 'mejor ajuste'
from scipy.stats import norm
mu, std = norm.fit(data)
p = norm.pdf(bins, mu, std)
plt.plot(bins, p, 'k', linewidth=2)

plt.title("Histograma con Curva de Distribución")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()
```

# Límites de Error

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
error = np.random.rand(10) * 0.5

plt.errorbar(x, y, yerr=error, fmt='-o')
plt.title('Ejemplo de Límites de Error')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

## Múltiple Figuras y Ejes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

## Escalas diferentes

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
